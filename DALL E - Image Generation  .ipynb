```python
import openai
import urllib.request
import cv2
from PIL import Image
import  matplotlib.pyplot as plt
```

```python
openai.api_key = 'API_KEY'
```

```python
img_res = openai.Image.create(
    prompt = 'praying in temple',
    n = 1,
    size = '512x512'
)
```

```python
img_res
```

```python
img_link = img_res['data'][0]['url']
img_link
```

```python
urllib.request.urlretrieve(img_link,"al_img.png")
```

```python
img  = cv2.imread('al_img.png')
```

```python
plt.imshow(img)
```
